```python
import os
from PIL import Image
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (10,5))
plt.suptitle("Image") # 图像窗口名称
img1 = Image.open('pcb.png')
plt.subplot(1,2,1)
plt.title('pcb') # 图像题目
plt.imshow(img1)
plt.axis('off') # 关掉坐标轴为 off
img2 = Image.open('LSTM2.jpg')
plt.subplot(1,2,2)
plt.title('LSTM') # 图像题目
plt.imshow(img2)
plt.axis('off') # 关掉坐标轴为 off
plt.show()
```

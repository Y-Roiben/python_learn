## plt.xticks(ticks=None, labels=None, **kwargs)

ticks:此参数是xtick位置的列表。和一个可选参数。如果将一个空列表作为参数传递，则它将删除所有xticks

labels:此参数包含放置在给定刻度线位置的标签。它是一个可选参数。

**kwargs:此参数是文本属性，用于控制标签的外观。

```python
x = np.linspace(0,10,100)
plt.plot(x, x*x)
```

##  显示x轴刻度从1到5，同时改变对应文本标签，并且让文本标签旋转一定角度（比如说30°）

```python
x = np.linspace(0,10,100)
plt.plot(x, x*x)
a = [1,2,3,4,5]
labels = ['A', 'B', 'C', 'D','E']
plt.xticks(a,labels,rotation = 30)
plt.show()
```

### 不想显示x轴刻度，则传入空列表即可

```python
x = np.linspace(0,10,100)
plt.plot(x, x*x)
a = [1,2,3,4,5]
plt.xticks([])
plt.show()
```

### 绘图 x行 y列第z个图 subplots( x行 ， y列， 第z个图 )

subplot(\*, \*, \*)

#### subplots(nrows=1, ncols=1, *, sharex=False, sharey=False, squeeze=True, subplot_kw=None, gridspec_kw=None, **fig_kw)

nrows：默认为 1，设置图表的行数。

ncols：默认为 1，设置图表的列数。

sharex、sharey：设置 x、y 轴是否共享属性，默认为 false，可设置为 'none'、'all'、'row' 或 'col'。 False 或 none 每个子图的 x 轴或 y 轴都是独立的，True 或 'all'：所有子图共享 x 轴或 y 轴，'row' 设置每个子图行共享一个 x 轴或 y 轴，'col'：设置每个子图列共享一个 x 轴或 y 轴。

squeeze：布尔值，默认为 True，表示额外的维度从返回的 Axes(轴)对象中挤出，对于 N*1 或 1*N 个子图，返回一个 1 维数组，对于 N*M，N>1 和 M>1 返回一个 2 维数组。如果设置为 False，则不进行挤压操作，返回一个元素为 Axes 实例的2维数组，即使它最终是1x1。

subplot_kw：可选，字典类型。把字典的关键字传递给 add_subplot() 来创建每个子图。

gridspec_kw：可选，字典类型。把字典的关键字传递给 GridSpec 构造函数创建子图放在网格里(grid)。

**fig_kw：把详细的关键字参数传给 figure() 函数。

```python
x = np.array([1,3,6,7,8])
y = np.array([3,9,8,1,10])

plt.subplot(1,3,1)
plt.plot(x,y)
plt.title("w")

plt.subplot(1,3,2)
plt.plot(x)
plt.title("x", family = 'KaiTi')

plt.subplot(1,3,3)
plt.plot(y)
plt.title("y")
plt.show()
```

#### 创建散点图 scatter()
x，y：长度相同的数组，也就是我们即将绘制散点图的数据点，输入数据。

s：点的大小，默认 20，也可以是个数组，数组每个参数为对应点的大小。

c：点的颜色，默认蓝色 'b'，也可以是个 RGB 或 RGBA 二维行数组。

marker：点的样式，默认小圆圈 'o'。

cmap：Colormap，默认 None，标量或者是一个 colormap 的名字，只有 c 是一个浮点数数组的时才使用。如果没有申明就是 image.cmap。

norm：Normalize，默认 None，数据亮度在 0-1 之间，只有 c 是一个浮点数的数组的时才使用。

vmin，vmax：：亮度设置，在 norm 参数存在时会忽略。

alpha：：透明度设置，0-1 之间，默认 None，即不透明。

linewidths：：标记点的长度。

edgecolors：：颜色或颜色序列，默认为 'face'，可选值有 'face', 'none', None。

plotnonfinite：：布尔值，设置是否使用非限定的 c ( inf, -inf 或 nan) 绘制点。

**kwargs：：其他参数。

```python
x = np.random.uniform(0, 5, 10)
y = np.random.uniform(0, 100, 10)
```

```python
plt.scatter(x,y)
plt.show()
```

# 颜色图
#### 使用带颜色图值的关键字参数，cmap指定颜色图，viridis,pink,hvs是matplotlib中内置颜色图之一

颜色条参数值可以是以下值：https://www.runoob.com/matplotlib/matplotlib-scatter.html

使用 plt.colorbar() 方法显示颜色条

```python
x = np.array([3,2,6,21,12,20,15,18,9,17,7])
y = np.array([33,22,60,21,40,30,40,18,34,50,37])
colors  =np.linspace(0,100,11,endpoint = False)
```

```python
plt.scatter(x,y,c=colors,cmap = 'pink')
plt.colorbar()
plt.show()
```

```python
plt.scatter(x,y,c=colors,cmap = 'hsv')
plt.colorbar()
plt.show()
```

#### 颜色， 尺寸，透明度
    alpha表示透明度

```python
x = np.random.randint(100,size=(100))
y = np.random.randint(100,size=(100))
colors = np.random.randint(100,size=(100))
sizes = 10*np.random.randint(100,size=(100))
plt.scatter(x, y, c=colors,s=sizes,alpha = 0.5, cmap = 'viridis')
plt.colorbar()
plt.title('散点图',size=20)
plt.show()
```

# 1创建柱状图

.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)

x：浮点型数组，柱形图的 x 轴数据。

height：浮点型数组，柱形图的高度。

width：浮点型数组，柱形图的宽度。

bottom：浮点型数组，底座的 y 坐标，默认 0。

align：柱形图与 x 坐标的对齐方式，'center' 以 x 位置为中心，这是默认值。 'edge'：将柱形图的左边缘与 x 位置对齐。要对齐右边缘的条形，可以传递负数的宽度值及 align='edge'。

**kwargs：：其他参数。

```python
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 6,11,17])
plt.bar(x,y,color = 'red')
plt.show()
```

```python
plt.barh(x,y,color='#4caf50',label = 'x-y')
plt.legend()
plt.show()
```

#### 条形宽高

```python
plt.bar(x,y,width = 0.2,label = 'x-y')
plt.legend()
plt.show()
```

```python
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 8,17,11])
plt.barh(x,y,height=0.2,color = 'pink')
plt.show()
```

## 饼状图  .pie()

 startangle(起始角度） = 0\90\180\270

```python
y = np.array([25,15,30,20,10])
x = ['a','b','c','d','e']
plt.pie(y, labels = x, startangle=0)
plt.show()
```

#### Explode 让某一块饼脱颖而出

```python
y = np.array([25,15,30,20,10])
x = ['a','b','c','d','e']
myexplode = [0,0.1,0,0,0]# 第二部分突出显示，值越大，距离中心越远
plt.pie(y,labels = x, explode = myexplode)
plt.legend()
plt.show()
```

#### 设置阴影 颜色 图例legend()

```python
y = np.array([25,15,30,20,10])
x = ['a','b','c','d','e']
myexplode = [0,0.1,0,0,0]
mycolors = ['red', 'black','blue','hotpink','#4caf55']
plt.pie(y,labels = x, explode = myexplode, shadow = True, colors = mycolors)
plt.legend(title = 'abcd',loc = 'upper left')
plt.show()
```

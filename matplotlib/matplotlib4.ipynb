#### 1创建柱状图

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array(['A', 'B', 'C', 'D'])
```

```python
y = np.array([3, 6,11,17])
```

```python
plt.bar(x,y,color = 'red')
plt.show()
```

#### 水平柱状图

```python
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 6,11,17])
```

###### 颜色设置支持十六进制，如（#4caf50）

```python
plt.barh(x,y,color='#4caf50')
plt.show()
```

#### 条形宽高

```python
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 8,17,21])
```

```python
plt.bar(x,y,width = 0.2)
plt.show()
```

```python
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 8,17,11])
plt.barh(x,y,height=0.1,color = 'pink')
plt.show()
```

#### 直方图 hist()

```python
x = np.random.normal(170,10,250)
```

```python
plt.hist(x)
plt.show()
```

#### 创建饼 pie() startangle(起始角度） = 0\90\180\270

```python
y = np.array([25,15,30,20,10])
x = ['a','b','c','d','e']
```

```python
plt.pie(y, labels = x, startangle=0)
plt.show()
```

#### Explode 让某一块饼脱颖而出

```python
y = np.array([25,15,30,20,10])
x = ['a','b','c','d','e']
myexplode = [0,0.1,0,0,0]
plt.pie(y,labels = x, explode = myexplode)
plt.show()
```

#### 设置阴影 颜色 图例legend()

```python
y = np.array([25,15,30,20,10])
x = ['a','b','c','d','e']
myexplode = [0,0.1,0,0,0]
mycolors = ['red', 'black','blue','hotpink','#4caf55']
plt.pie(y,labels = x, explode = myexplode, shadow = True, colors = mycolors)
plt.legend(title = 'abcd')
plt.show()
```
